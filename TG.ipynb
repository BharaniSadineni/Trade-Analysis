```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import FileLink, display
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.drawing.image import Image
import os

# Define the correct file path
file_path = "E:/UE Sose2024/Projects/Data Science and Business/Assignment/XL partners all/Partners all 2023-2017 XL.xlsx"

# List of columns to read
columns_to_read = [
    'Period', 'ReporterDesc', 'PartnerDesc', 'CmdCode', 'QtyUnitAbbr',
    'Qty', 'AltQtyUnitAbbr', 'AltQty', 'NetWgt', 'PrimaryValue'
]

# Load the specified columns from the Excel file
data = pd.read_excel(file_path, usecols=columns_to_read)

# Convert AltQty from grams to kilograms where AltQtyUnitAbbr is 'g'
mask = data['AltQtyUnitAbbr'] == 'g'
data.loc[mask, 'AltQty'] = data.loc[mask, 'AltQty'] / 1000
data.loc[mask, 'AltQtyUnitAbbr'] = 'kg'

# Convert PrimaryValue to millions
data['PrimaryValue'] = data['PrimaryValue'] / 1e6

# Sort the DataFrame by ReporterDesc, Period, CmdCode, and PartnerDesc
sorted_data = data.sort_values(by=['ReporterDesc', 'Period', 'CmdCode', 'PartnerDesc'])

# Remove the QtyUnitAbbr and AltQtyUnitAbbr columns
sorted_data = sorted_data.drop(columns=['QtyUnitAbbr', 'AltQtyUnitAbbr'])

# Rename the PrimaryValue column to indicate the unit is in millions
sorted_data = sorted_data.rename(columns={'PrimaryValue': 'PrimaryValue (millions)'})

# Filter data for Togo and Zimbabwe
togo_data = sorted_data[sorted_data['ReporterDesc'] == 'Togo']
zimbabwe_data = sorted_data[sorted_data['ReporterDesc'] == 'Zimbabwe']

# Function to calculate descriptive statistics
def descriptive_stats(df, group_by_cols):
    stats = df.groupby(group_by_cols).agg({
        'Qty': ['mean', 'median', 'std', 'min', 'max'],
        'AltQty': ['mean', 'median', 'std', 'min', 'max'],
        'NetWgt': ['mean', 'median', 'std', 'min', 'max'],
        'PrimaryValue (millions)': ['mean', 'median', 'std', 'min', 'max']
    }).reset_index()
    # Flatten MultiIndex columns
    stats.columns = [' '.join(col).strip() for col in stats.columns.values]
    return stats

# Calculate descriptive statistics for Togo and Zimbabwe
togo_stats = descriptive_stats(togo_data, ['PartnerDesc', 'CmdCode'])
zimbabwe_stats = descriptive_stats(zimbabwe_data, ['PartnerDesc', 'CmdCode'])

# Display statistics
print("Togo Descriptive Statistics:")
print(togo_stats)

print("\nZimbabwe Descriptive Statistics:")
print(zimbabwe_stats)

# Function to save statistics and plots in the same Excel sheet
def save_stats_and_plots(togo_stats, zimbabwe_stats, output_file_path):
    with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:
        workbook = writer.book
        sheet = workbook.create_sheet(title="Descriptive Statistics")
        
        # Write statistics to the sheet
        start_row = 1
        
        for stats, country in [(togo_stats, 'Togo'), (zimbabwe_stats, 'Zimbabwe')]:
            for metric in ['Qty', 'AltQty', 'NetWgt', 'PrimaryValue (millions)']:
                sheet.cell(row=start_row, column=1, value=f"{country} {metric} Statistics")
                metric_stats = stats[['PartnerDesc', 'CmdCode', f'{metric} mean', f'{metric} median', f'{metric} std', f'{metric} min', f'{metric} max']]
                for r_idx, row in enumerate(dataframe_to_rows(metric_stats, index=False, header=True)):
                    for c_idx, value in enumerate(row, 1):
                        sheet.cell(row=start_row + r_idx + 1, column=c_idx, value=value)
                start_row += len(metric_stats) + 3
                
                # Generate plot for each metric
                plt.figure(figsize=(10, 6))
                metric_stats.plot(kind='bar', x='PartnerDesc', y=f'{metric} mean', title=f'{country} {metric} Mean')
                plot_path = f"{country}_{metric}_mean_plot.png"
                plt.savefig(plot_path)
                plt.close()
                
                # Insert plot into Excel sheet
                img = Image(plot_path)
                img.anchor = f"A{start_row}"
                sheet.add_image(img)
                start_row += 20
        
        # Add a report summary
        report_text = """Report Summary:
        This analysis provides descriptive statistics for the trade data of Togo and Zimbabwe from 2017 to 2023.
        Each metric (Qty, AltQty, NetWgt, PrimaryValue) is analyzed separately, showing mean, median, standard deviation, minimum, and maximum values.
        The corresponding plots are included for better visualization.
        """
        sheet.cell(row=start_row, column=1, value=report_text)
    
    print(f"Descriptive statistics and plots saved to {output_file_path}")

# Save statistics and plots in the same Excel sheet
output_file_path = os.path.join(os.getcwd(), "Descriptive_Statistics_Togo_Zimbabwe.xlsx")
save_stats_and_plots(togo_stats, zimbabwe_stats, output_file_path)

# Check if the file was created
if os.path.exists(output_file_path):
    print(f"Descriptive statistics saved to {output_file_path}")
    # Provide a link to download the file
    display(FileLink(output_file_path))
else:
    print(f"Failed to save the file at {output_file_path}")
```
